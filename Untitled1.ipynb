```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
input ="hi good evening everyone"
output1 = word_tokenize( input)
print(output1)
```

```python
input = "hi good evening everyone. nice to meet you guys"
output2 = sent_tokenize(input)
print(output2)
```
